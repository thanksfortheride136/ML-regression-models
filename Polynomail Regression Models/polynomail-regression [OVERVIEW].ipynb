# Polynomial Regression: Beginner Walkthrough

## What is this notebook about?

In this notebook, we are learning how to use **polynomial regression**.

Simple linear regression is useful when data follows a mostly straight-line pattern.

But sometimes data is curved.

Polynomial regression helps us fit curved data by creating new input features like:

- x
- x²
- x³

The important idea is:

> Polynomial regression is still linear regression, but we give the model extra features that help it fit curves.

# 1. What is a polynomial?

A polynomial is a math expression that uses powers of a variable.

For example:

- x
- x²
- x³

In Python:

```python
x
x**2
x**3
```
