A multiple linear regression model uses several input variables to predict one output variable. In this example, the model uses house size, number of bedrooms, number of floors, and house age to predict house price.

## 2. Import Libraries

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## 3. Check That SciKit-Learn Works

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
print("scikit-learn is working")
```

## 4. Create or Import Training Data
Here you can feed data from other csv files, lists etc

```python
X_train = np.array([
    [2104, 5, 1, 45],
    [1416, 3, 2, 40],
    [852, 2, 1, 35],
    [1940, 4, 2, 20],
    [1200, 3, 1, 30],
    [2500, 4, 2, 10],
    [1800, 3, 2, 25],
    [3000, 5, 2, 8],
    [1100, 2, 1, 50],
    [1600, 3, 1, 35],
    [2750, 4, 2, 15],
    [950, 2, 1, 60],
    [2200, 4, 2, 18],
    [1350, 3, 1, 42],
    [3200, 5, 3, 5]
])

# Each row represents one house/example.
# Each column represents one feature: Size, Bedrooms, Floors, Age.
```

```python
y_train = np.array([
    460,
    232,
    178,
    410,
    245,
    540,
    365,
    680,
    190,
    310,
    590,
    165,
    475,
    255,
    725
])
# y_train contains the correct output values.
# In this case, each value is the price of the house in thousands of dollars.
# Example: 460 means $460,000.
```

## 5. Create the Model

```python
model = LinearRegression()
```

## 6. Train the Model and Find the Weights

The model finds the best-fit equation by choosing weights and a bias that make the prediction error as small as possible.

```python
model.fit(X_train, y_train) # finds the best-fit equation

#Printing info below
feature_names = ["Size", "Bedrooms", "Floors", "Age"]

for feature, weight in zip(feature_names, model.coef_):
    print(f"{feature} weight: {weight:.2f}")

print(f"Bias: {model.intercept_:.2f}")
```

## 7. Put Data into a DataFrame

```python
feature_names = ["Size", "Bedrooms", "Floors", "Age"]

df = pd.DataFrame(X_train, columns=feature_names)

df["Price"] = y_train

df
```

## 8. Graph on Feature Against Price

```python
sns.regplot(data=df, x="Size", y="Price")
plt.title("House Size vs. Price")
plt.xlabel("Size in Square Feet")
plt.ylabel("Price in Thousands of Dollars")

plt.show()
```

## 9. Graph all Features Against Price

```python
sns.pairplot(
    df,
    x_vars=["Size", "Bedrooms", "Floors", "Age"],
    y_vars="Price",
    kind="reg"
)

plt.show()
```

## 10. Make a Prediction

```python
# New house: 1500 sqft, 4 bedrooms, 1 floor, 40 years old
new_house = [[1500, 4, 1, 40]]

prediction = model.predict(new_house)

estimated_price = prediction[0] * 1000

print(f"Estimated House Price: ${estimated_price:,.2f}")
```
