## Simple Linear Regression ML Model - Car Stopping Distance Example
This model can make predictions for new speeds, but predictions are most reliable near the range of the training data. Since the training data ranges from 4–25 mph, predictions far outside that range are extrapolations and may be unreliable.

### 1. Imports

```python
%pip install seaborn
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

### 2. Supply Input Training Data
Data must be cleaned prior 

```python
x_train = np.array([
    4, 4, 7, 7, 8, 9, 10, 10, 10, 11,
    11, 12, 12, 12, 12, 13, 13, 13, 13, 14,
    14, 14, 14, 15, 15, 15, 16, 16, 17, 17,
    17, 18, 18, 18, 18, 19, 19, 19, 20, 20,
    20, 20, 20, 22, 23, 24, 24, 24, 24, 25
    ])
y_train = np.array([
    2, 10, 4, 22, 16, 10, 18, 26, 34, 17,
    28, 14, 20, 24, 28, 26, 34, 34, 46, 26,
    36, 60, 80, 20, 26, 54, 32, 40, 32, 40,
    50, 42, 56, 76, 84, 36, 46, 68, 32, 48,
    52, 56, 64, 66, 54, 70, 92, 93, 120, 85
])
print(f"X Training Data: {x_train}")
print(f"Y Training Data: {y_train}")
```

### 3. Calculate Number of Training Examples
This is denoted by "m", each X has a corresponding Y value, which is why you only need to do this on X.

```python
m = len(x_train)
print(f"There are {m} (x, y) training examples in this dataset.")
```

### 4. Access Individual Training Examples

```python
i = 0 #first index
x_i = x_train[i]
y_i = y_train[i]
print(f"X: {x_i}, Y: {y_i}")
```

### 5. Create a Scatter Plot Using Seaborn to Visualize Data

```python
sns.regplot(x=x_train, y=y_train)
plt.title("Car Speed vs. Stopping Distance")
plt.xlabel("Speed (mph)")
plt.ylabel("Stopping Distance (ft)")
plt.show()
```

Since this is simple linear regression, we are using one input feature, speed, to predict one target value, stopping distance. The model uses the equation:

predicted_y = w * x + b

Where w is the weight/slope and b is the bias/y-intercept.

### 6. Find the Weight and Bias

```python
w, b = np.polyfit(x_train, y_train, 1) #finds the weight (slope) and bias (y intercept). 1 is just mx+b shape 
```

```python
print(w, b)
```

### 7. Prediction

```python
x_i = 21
stopping_distance = w * x_i + b    
print(f"Speed: {x_i} mph")
print(f"Predicted stopping distance: {stopping_distance:.2f} ft")
```
